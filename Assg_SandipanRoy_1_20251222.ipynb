```python
# List is one of the derived datatypes in python. It is capable of storing more than one value
# all together. List can be created -
#   1. by instantiating a variable with the list() method
#   2. by assigning a variable with []
# We can either left them with no values or at the time of creation only we can assign values.
# List are mutable in nature, i.e., we can add, update or remove values inside a list object/variable.
# Another feature of list is it can store duplicate values as well.

lst = list()
print(type(lst), lst)

lst = list([1,"God",3.14, True])
print(type(lst), lst)

lst = []
print(type(lst), lst)

lst = [1,2,3, "God", 3.14, False, -10]
print(type(lst), lst)
```

```python
lst01 = list(range(10))
lst01
```

```python
# In this section we have used 3 important methods and functions for list operation.
#     1. len():- the len() function calculates the number of elements or values which
#        are stored within a variable
#     2. append():- the append() method adds one element/value at the end of the list
#     3. insert():- the insert() method inserts a value in the given index within a list.
#        However, the given index should be proper, else it will throw an error.

lst1 = []
print(len(lst1), "List:", lst1)

lst1.append(1)
print(len(lst1), "List:", lst1)
lst1.append(2)
lst1.append(3)
print(len(lst1), "List:", lst1)
lst1.insert(0, 0)
print(len(lst1), "List:", lst1)
```

```python
# In this section, we have used another important method associated with list- remove().
# The remove() method removes the first occurance of the given value from the list. If
# a list is having multiple duplicate value, then the remove function will only remove
# its first occurance. Remaining values will remain untouched.

lst2 = [1,2,3,4,5,6]
print(type(lst2), "List:", lst2)
lst2.append(2)
print("List:", lst2)

# Section:- List.remove(value)
lst2.remove(2)
print("List:", lst2)

lst2.remove(3)
print(lst2)
```

```python
# In this section we have used another useful important method of list- pop().
# The pop() method removes items from the last position of the list and returns the removed value.
# It performs on reverse indexing over the list. It accepts one optional parameter
# value for this reverse index and it's default value is -1.

lst2.append(3)
print("List:", lst2)

print("Removed value:", lst2.pop())
print("List:", lst2)

print("Removed value:", lst2.pop(-2))
print("List:", lst2)
```

```python
# In this section we have used another important method of list - sort(); which is
# capabple of ordering the list items. The ordering takes in-place, i.e., the list
# itself is ordered and the method returns None.

print(f"Before sorting: {lst2}")
lst2.sort()
print(f"After sorting: {lst2} \n")
```

```python
# In this section we have used one important function- sorted(). This function is
# capable of sorting any iterable data and return a new ordered data rather than
# modifying the original iterable data.

lst2 = [1,2,3,4,5,6, -1,-2,-3,-4,-5,-6,0]
lst3 = sorted(lst2)
print(f"Original list: {lst2}")
print(f"New Sorted list: {lst3}")
```

```python
lst4 = [0, 1, 2, 3, 4, 5, 6]
lst5 = [-6, -5, -4, -3, -2, -1]

# Indexing in the python programming follows 2 thumb rules:
#     1. The index of first element of an iterable ordered data is zero and then
#        it increases by one. In other words it follows a simple mathematical formula:
#               index = Position - 1
#        So when we are refering to the first element of a list, then
#            Position = 1
#            Index = Position - 1 = 1-1 = 0
#
#     2. If we are willing to traverse any iterable ordered data from the back which
#        is also refered as reverse indexing then the index value starts as:
#            index = Position from last * -1
#        So when we are refering the last element of the list, which is basically
#        the first element from the last then:
#            Position from the end = 1
#            index = Position from the end * -1 = 1 * (-1)= -1
#        Similarly the second element index value in this particular scenario will be -2.
#
# In the above we can see 2 list variables. For the understanding purpose only I've
# assigned values based on the indexing logic of python programming.
# "lst4"- represents the forward index positioning logic and "lst5"- represents
# the reverse index logic.

# Slicing in iterable ordered data works on
#    [starting index : ending index : step size]
# Operationally we can define it like- first it will pick up the element in the
# as mentioned as "starting index" position element; then it will increment the index
# value by the "step size" and will move forward till it reaches the "ending index" - 1 th
# position.
# Note: The "ending index" value always excludes that index position element
#       and the "step size" always represents the incremental factor, which determines
#       which element to pick up next

#-------------------------------------------------------------------------------------
# Slicing by forward index
#-------------------------------------------------------------------------------------

print("Forward Indexing")
print("----------------------------------------------------\n")
# This slicing will fetch data from 0th index to 3rd index without skipping any element
# i.e, it will first pickup the 0 then 1 then 2 then 3
print(lst4[0:4])
# This slicing will fetch data from the starting index to 3rd index (4-1) index, i.e.,
# it will first pickup the 0 then 1 then 2 then 3
print(lst4[:4])
# This slicing will pick up from the 1st index till the last element of the index without
# skipping any element in between.i.e, the output will be 0,1,2,3,4,5,6.
print(lst4[1::1])
# This slicing will pick up from the 0th index till the last index and will not skip any
# index position. In other words it will print all the elements in the list
print(lst4[::])
# This slicing will start from 0th index and increment the index positional value by 2
# thus skipping every alternative numbers. The output will be 0,2,4,6
print(lst4[::2])

#-------------------------------------------------------------------------------------
# Slicing by reverse indexing
#-------------------------------------------------------------------------------------

# In case of reverse indexing we always has to mention the step size which will be in
# this case a  negative number else it will through an error.
# If we want to fetch each and every item from the list in reverse order then the step
# size will be -1. Similarly we want to fetch alternative values then it will be -2.
# For starting index as well as ending index they will also negative values.

print("----------------------------------------------------\n")
print("Reverse Indexing")
print("----------------------------------------------------\n")
print(lst5[-1::-1])
print(lst5[-1:-4:-1])
print(lst5[-1::-2])
print(lst5[::-1])
```

</ul>

```python
n = int(input("Enter a number: "))

if n <= 2 and n > 0:
  if n%2==0:
    print(f"{n} is an even number")
  else:
    print(f"{n} is an odd number")
else:
  if n%2==0:
    print(f"{n} is an even number")
  else:
    print(f"{n} is an odd number")
    up_limit = n//2 + 1
    isPrime = True
    for i in range(2,up_limit):
      if n%i == 0:
        isPrime = False
        break
      else:
        continue
  if isPrime:
    print(f"{n} is a Prime number.")
```

```python
def getLargestNumber(lst):
  n = len(lst)
  if n == 1:
    return lst[0]
  else:
    max_num = lst[0]
    for num in lst[1:]:
      if num > max_num:
        max_num = num
    return max_num


if __name__=="__main__":
  is_valid_data =False

  while not is_valid_data:

    entered_values = input("Enter numbers separated by space:")

    num_lst = [int(x) for x in entered_values.split() if x.isnumeric()]

    if len(num_lst) == 0 or len(num_lst) < len(entered_values.split()):
      print("Please enter numbers only separated by space.")
    else:
      is_valid_data = True

  print(getLargestNumber(num_lst))
```

```python
def element_frequency(string):
  """ The purpose of this fuction will be to
  return the character frequency in the given string """
  s1=str(string.lower()).replace(" ","")
  char_dict = dict()
  for c in s1:
    if char_dict.get(c)==None:
      char_dict[c] = 1
    else:
      char_dict[c] += 1

  for k,v in char_dict.items():
    print(f"'{k}':-> {v} time(s)")

if __name__=="__main__":
  entered_values = input("Please enter the elements:")
  element_frequency(entered_values)

```

```python
def fibonacci_series(num):
  fib_lst = []
  a = 0
  b = 1

  fib_lst.append(a)

  if num == 1:
    pass
  elif num > 1:
    n = 1
    while n < num:
      fib_lst.append(b)
      a, b = b, a + b
      n += 1

  return fib_lst

if __name__ == "__main__":
  fib_lst = []
  try:
    num = int(input("Please enter a number: "))

    if num > 0:
      fib_lst = fibonacci_series(num)

    print(f" First {num} elements in Fibonacci Series: {fib_lst}")
  except:
    print("Please enter a valid number.")
```
