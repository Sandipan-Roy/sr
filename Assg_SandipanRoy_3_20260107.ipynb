```python
arr1 = np.array([0, 1, 2, 3, 4, 5])

# Shape of the array:
# Returns a tuple representing the dimensions of the array.
print("Shape of arr1:", arr1.shape)
# Size of the array:
# Returns the total number of elements in the array.
print("Size of arr1:", arr1.size)
# Number of dimensions:
# Returns the number of dimensions (axes) of the array.
print("Number of dimensions of arr1:", arr1.ndim)
# Data type of the elements of the array:
# Returns the data type of the elements stored in the array.
print("Data type of arr1:", arr1.dtype)
```

```python
# Numpy arrays are homogeneous in nature, meaning all elements in the array must be of the same data type.
# If you try to create an array with mixed data types, NumPy will automatically convert all elements to a
# common data type that can accommodate all values.
# For example, if you mix integers and floats, NumPy will convert all integers to floats to maintain consistency.
# If we mix datatypes that cannot be converted to a common type, NumPy will default to using the most general type,
# which is usually a string (object) type.
# This allows the array to hold elements of different types, but it comes at the cost of performance and memory efficiency.
# In any case if numpy fails to convert, it will raise a ValueError indicating that the conversion is not possible.

arr2 = np.array([1, 2.5, 3, 4.0, 5])
print("arr2:", arr2)
print("Data type of arr2:", arr2.dtype)

arr3 = np.array([1, 'two', 3.0, True])
print("arr3:", arr3)
print("Data type of arr3:", arr3.dtype)

# The above code demonstrates how NumPy handles arrays with mixed data types by converting them to a common type.

# However if we uncomment the below line, it will raise an error because of it's 4th element which is in
# in conflict of the above explained property- conversion into common data type and maintaining it's homogeneous
# nature.
# arr4 = np.array([1, 2, 'three', [4, 5]])
```

```python
# Numpy allows us to concatenate 2 arrays, the pre-requisite of this concatenation
# operation is the dimensionality of the arrays has to match over the given axis,
# else it will through ValueError.

arr10 = np.array([[4,5,6]])
print(f"Array 10 Shape: {arr10.shape} Array 10:")
print(arr10)
print("="*40)
arr11 = np.array([[1,2,3],[7,8,9],[10,11,12]])
print(f"Array 11 Shape: {arr11.shape} Array 11:")
print(arr11)
print("="*40)

print("Concatenation")
print("="*40)
arr12 = np.concatenate((arr10, arr11), axis= 0)
print("Array 10 concatenate Array 11- Axis= 0 or Row-Wise or Vertical Level")
print(arr12)

# For the above scenario, the axis which we have mentioned is 0 i.e., Row-wise
# or concatenation at vertical level. So for this operation to be successfull
# the dimensionality at the row-level has to match.
# However if we will try to concatenate the same array with axis=1 i.e.,
# column-wise or caoncatenate at horizontal-level it will through an error due
# to mismatch of dimensionality at the column level.

print("="*40)
print()
arr13 = np.array([[4],[5],[6]])
print(f"Array 13 Shape: {arr13.shape} Array 13: ")
print(arr13)
print("="*40)
print(f"Array 11 Shape: {arr11.shape} Array 11:")
print(arr11)
print("="*40)

print("Concatenation")
print("="*40)
arr14 = np.concatenate((arr13, arr11), axis= 1)
print("Array 14 concatenate Array 11- Axis= 1 or Column-Wise or Horizontal Level")
print(arr14)


# In simple laymans way how we can explain it w.r.t. the above code is when
# concatenation is happening with axis= 0 then it is actually considering row-wise
# element count, matching it and increase the size of the array vertically or
# increasing the row-count; whereas when axis=1 then it is considering column wise
# element count, matching it and increasing the size of array horizontally or
# increasing the column-count.
```

```python
# np.concatenate, np.hstack, and np.vstack all combine arrays, but they differ
# in their primary use cases and how they handle array dimensions.
# np.concatenate is the general-purpose function, while np.hstack and np.vstack
# are specialized, convenience wrappers for horizontal and vertical stacking, respectively.

arr15 = np.hstack((arr13, arr11))
arr16 = np.vstack((arr10, arr11))

# 'arr15' and 'arr14' operation are same and will give the same output. Similarly,
# 'arr16' and 'arr12' operation are same and will give the same output.

print(arr15)
print(arr14)
print("="*40)
print(arr16)
print(arr12)
```

```python
# In Pandas, the df.fillna() method is used to replace missing values
# (represented as NaN) with specific data. This is a critical step in data
# cleaning to ensure datasets are complete for analysis or machine
# learning models.
# Here, we are filling the missing values with 0


statDF["sepal_length"].fillna(0, inplace=True)
statDF
```

```python
def normalize_numerical_columns(df, columns):
    """
    Normalize specified numerical columns in a DataFrame using Min-Max scaling.

    Parameters:
    df (pd.DataFrame): The input DataFrame containing the data to be normalized.
    columns (list): A list of column names to be normalized.

    Returns:
    pd.DataFrame: A DataFrame with the specified columns normalized.
    """
    df_normalized = df.copy()
    for col in columns:
        min_val = df_normalized[col].min()
        max_val = df_normalized[col].max()
        df_normalized[col] = (df_normalized[col] - min_val) / (max_val - min_val)
        #print(f"Normalized column: {col}")
        #print(df_normalized[col].head())
    return df_normalized
```

**Comparission:**


*   Our original dataframe(irisDF) memory usage was 6.0+ KB
*   After One-Hot-Encoding, the resultant new dataframe(iris_DF1) after dropping the categorical column memory usage increased to 8.3 KB
*   After Label-encoding using .astype(), the resultant new dataframe(irisDF2) memory usage dropped down to 5.0 KB

**Observation:**


*   In earlier cases we have seen, when the no of categories within a feature is limited to 2 and when used One-Hot-Encoding the memory usage dropped down.
*   However, when the no of categories within a feature is more than 2, in that scenario limiting it within a single column is not possible using OHE which resulted in the feature engineering of columns depending on the number of categories (in this case, we are having 3 categories so our resultant new features is 3). This increased the memory usage of the dataframe.
*   Again, when we used Label-Encoding on the same dataframe, since the resultant feature-engineering is limited to only 1 feature/column, so the memory usage dropped.

**Conclussion:** One-Hot-Encoding will be useful in case the number of categories within a column is limited whereas we have to use Label-Encoding in case where number of categories is more.
