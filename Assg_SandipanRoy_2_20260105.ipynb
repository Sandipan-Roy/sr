We can install any packages in python with the help of "pip" command. This command is an acronym which stands for "pip installs packages". This command is capable of handling hierarchical dependency resolution in package installation. Suppose we want to install a package "xyz", which internally is dependant on an another package "abc1" which in turn is dependant on an another package "abc2" and so on. So for installing package, the only thing which we have to do is to execute the below command:

          > pip install xyz
            
    The dependency resolver within the pip command will resolve all the heirarchical dependency tree and will install all the necessary packages along with their compatible verssion.

This is where the full form of "pip" which is "pip installs packages" fits in appropriately, where pip in "pip installs packages" itself stands for "pip install packages".

Another way of writing pip command depending on OS and/or IDE is:

    > !pip install xyz
            
Or

    > %pip install xyz
            

So, to install "numpy" package in our environment, we need to execute the command:

    > pip install numpy
            It will install the latest verssion of numpy package in our environment which is compatible with our present python verssion.

Similarly, to install "pandas" package in our environment, we need to execute the command:

    > pip install pandas
            It will install the latest verssion of pacndas package in our environment which is compatible with our present numpy package and python verssion.

So in other words, if we are going to install both numpy and pandas in our system, executing "pip" command for "pandas" only will achieve our objective. Since "pandas" package itself is built upon "numpy" package.
